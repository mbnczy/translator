```python
import pandas as pd
from utils.validation import length_check
from utils.postproc import concat
import logging

#data_name = 'cz_laws_onlytext'
#num_chunks = 10
#data_name = 'cz_bills_onlytext'
#num_chunks = 4
data_name = 'Szejm_scraped_2011-2023_bábelre_final'
num_chunks = 29

logger = logging.getLogger(f"{data_name}")
logger.setLevel(logging.INFO)
fh = logging.FileHandler(f"{data_name}.log")
fh.setLevel(logging.INFO)
fh.setFormatter(logging.Formatter('%(asctime)s - %(levelname)s - %(message)s'))
logger.addHandler(fh)
```

```python
length_check(
    chunks_folder = f"../data/tmp_{data_name}/chunks",
    out_folder = f"../data/tmp_{data_name}/out",
    num_chunks = num_chunks,
    logger = logger
)
```

```python
concat(
    out_folder = f"../data/tmp_{data_name}/out",
    result_name = f"../data/{data_name}_translated.csv",
    num_chunks = num_chunks,
    logger = logger
)
```
